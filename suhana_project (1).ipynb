```python

#funnel chart-top performing states
state_profit=data.groupby("State")["Sales Profit"].sum().sort_values(ascending=False)
state_profit_percent = (state_profit / state_profit.sum() * 100).round(1)
plt.figure(figsize=(6,8))
plt.barh(state_profit.index,state_profit.values)
plt.gca().invert_yaxis()
plt.title("Sales Funnel by State(Top to Bottom)")
plt.xlabel("Sales profit")
for i,(value,p)in enumerate (zip(state_profit.values,state_profit_percent.values)):
    plt.text(value,i,f"{p:.1f}%",va="center")
plt.show()
```

```python

import matplotlib.pyplot as plt

# Create the pivot table with 'Category' as index and 'Store Type' as columns
pivot = data.pivot_table(values="Sales Profit", index="Category", columns="Store Type", aggfunc="sum")

# Plot the stacked bar chart and capture the Axes object
ax = pivot.plot(kind="bar", stacked=True, figsize=(10, 7)) # Assign the plot object to 'ax'

# Set title and labels
plt.title("Stacked Bar chart: Sales Profit by Category and Store Type")
plt.xlabel("Category") # Add x-axis label
plt.ylabel("Sales Profit")
plt.xticks(rotation=45, ha='right') # Rotate x-axis labels for better readability

# Calculate percentages for labels within each stack (category total)
# Divide each cell by the sum of its row (total sales for that category)
pivot_percentages = pivot.div(pivot.sum(axis=1), axis=0) * 100

# Add labels to the bars
for i, container in enumerate(ax.containers):
    # container corresponds to a 'Store Type' (e.g., 'Owned' or 'Rented')
    store_type_col_name = pivot.columns[i]
    labels = [f'{val:.1f}%' for val in pivot_percentages[store_type_col_name]]
    ax.bar_label(container, labels=labels, label_type='center', fmt='%.1f%%')

plt.tight_layout() # Adjust layout to prevent labels/title from being cut off
plt.show()
```
